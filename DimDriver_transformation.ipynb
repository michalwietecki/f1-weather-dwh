```python
import pandas as pd

files = ['formula1_2020season_drivers.csv', 'formula1_2021season_drivers.csv', 'Formula1_2022season_drivers.csv', 'Formula1_2023season_drivers.csv', 'Formula1_2024season_drivers.csv']

merged_df = [pd.read_csv(file) for file in files]
merged_df = pd.concat(merged_df)
merged_df = merged_df.drop(columns=['No'])
merged_df = merged_df.drop(columns=['Podiums'])
merged_df = merged_df.drop(columns=['Number'])
merged_df = merged_df.drop(columns=['Points'])
merged_df = merged_df.drop(columns=['Grands Prix Entered'])
merged_df = merged_df.drop(columns=['World Championships'])
merged_df = merged_df.drop(columns=['Highest Race Finish'])
merged_df = merged_df.drop(columns=['Team'])
merged_df = merged_df.drop(columns=['Highest Grid Position'])

merged_df = merged_df.drop_duplicates()
```

```python
merged_df['DriverID'] = ['{:03d}'.format(i+1) for i in range(len(merged_df))]

# Przenie≈õ DriverID na poczƒÖtek kolumn
cols = ['DriverID'] + [col for col in merged_df.columns if col != 'DriverID']
merged_df = merged_df[cols]
```
