```python
print(results_2020.iloc[0,5])
print(results_2021.iloc[0,5])
print(results_2022.iloc[0,5])
print(results_2023.iloc[0,5])
print(results_2024.iloc[0,5])
```

```python
# 4NA7E1
def format_date(date):
    day, month, year = date.split()
    if len(day) == 1:
        day = "0" + day
    def switch_month(x):
        if x == "Jan":
            return "01"
        elif x == "Feb":
            return "02"
        elif x == "Mar":
            return "03"
        elif x == "Apr":
            return "04"
        elif x == "May":
            return "05"
        elif x == "Jun":
            return "06"
        elif x == "Jul":
            return "07"
        elif x == "Aug":
            return "08"
        elif x == "Sep":
            return "09"
        elif x == "Oct":
            return "10"
        elif x == "Nov":
            return "11"
        elif x == "Dec":
            return "12"
    month = switch_month(month)
    return day + "/" + month + "/" + year
```

```python
format_date(results_2020.iloc[0,5])
```

```python
results_2020["Race Date"] = results_2020["Race Date"].apply(format_date)
results_2021["Race Date"] = results_2021["Race Date"].apply(format_date)
```
